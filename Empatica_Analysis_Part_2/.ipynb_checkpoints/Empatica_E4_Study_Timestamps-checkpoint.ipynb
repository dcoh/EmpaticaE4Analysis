```python
import pandas as pd
import numpy as np
import csv
import glob
import datetime
import dateutil.parser as dp
```

```python
base_path = r"C:\Users\Dan Cohen\Desktop\Empatica_Data\ID PR025\responses" 
participant_id = "PR025"
```

```python
list_of_files = glob.glob(base_path + "/survey_responses*.csv")
```

https://stackoverflow.com/questions/1703546/parsing-date-time-string-with-timezone-abbreviated-name-in-python

```python
timezone_offsets = {'PST': -8*3600, 
                    'PDT': -7*3600,
                    'EDT': -4*3600,
                    'EST': 5*3600,
                    'MST': -7*3600,
                    'MDT': -6*3600,
                    'CDT': -5*3600,
                    'CST': -6*3600}
```

```python
for file in list_of_files:
    time_df = pd.read_csv(file)
    print(time_df['Scheduled Time'])
    
    utc_with_offsets = time_df.apply(lambda x: dp.parse(x['Scheduled Time'], tzinfos=timezone_offsets), axis=1)
    print(utc_with_offsets)
```
